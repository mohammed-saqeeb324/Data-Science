```python
#plot the boplot
sns.boxplot(x=df['Total Sales Value'])
plt.title('Box Plot To Visualize The Outliers')
plt.show()

# Identify Outliers Pragmatically
Q1=df['Total Sales Value'].quantile(0.25)
Q3=df['Total Sales Value'].quantile(0.75)
IQR=Q3-Q1 #interquartile range

print('First Quartile',Q1)
print('Third Quartile',Q3)
print('IQR',IQR)

#outlier Range
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

print('lower_bound')
print('upper_bound')

#Find Outliers
outliers=df[(df['Total Sales Value']<lower_bound) | (df['Total Sales Value']>upper_bound)]
print('outliers:',outliers)
```
