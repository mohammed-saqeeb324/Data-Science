## Task 2: Data Preprocessing
1.Encode categorical variables if necessary.
2.Split the dataset into training and testing sets.

```python
df.info()
```

```python
#apply the data transfromation technique label encoding on categorical data

from sklearn.preprocessing import LabelEncoder

label=LabelEncoder()


for i in range(0,23):

    df.iloc[:,i]=label.fit_transform(df.iloc[:,i])


df.head()
```

```python
#storing the result in other variable

dfcat=df.iloc[:,:23]
dfcat
```

```python
#applying data transformation using standardization technique

from sklearn.preprocessing import StandardScaler

scale=StandardScaler()

df_cat=df[['stalk_height','cap_diameter']]

df_cat=scale.fit_transform(df_cat)

df_cont=pd.DataFrame(df_cat,columns=['stalk_height','cap_diameter'])
df_cont
```

```python
#merging the two transformed data set into to a new dataset

df_new=pd.concat([dfcat,df_cont],axis=1)

df_new
```

## Data Partition

```python
#spliting the data into train and test

from sklearn.model_selection import train_test_split

x=df_new.drop('class',axis=1)
y=df_new['class']


xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=37)
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

## Task 4: SVM Implementation
1.Implement a basic SVM classifier using Python libraries like scikit-learn.
2.Train the SVM model on the training data.
3.Evaluate model performance on the testing data using appropriate metrics (e.g., accuracy, precision, recall, F1-score).

```python
ytrain= ytrain.astype(int)
```

```python
ytrain.isnull().sum()
```

```python
#implementing the svm classifier

from sklearn.svm import SVC

model=SVC(kernel='linear',C=5)

model.fit(xtrain,ytrain)

y_pred_train=model.predict(xtrain)
y_pred_test=model.predict(xtest)
```

```python
#calculating the accuracy score of the model

from sklearn.metrics import accuracy_score

ytest = ytest.astype(int)

accuracy=accuracy_score(ytrain,y_pred_train)
accuracytest=accuracy_score(ytest,y_pred_test)

print("accuracy train score:",accuracy)
print("accuracy test score:",accuracytest)
```

```python
#from classification report we can get the all the classification metrics

from sklearn.metrics import classification_report

print(classification_report(ytrain,y_pred_train))
```

```python
#calulating the all metrics mentioned in the document

from sklearn.metrics import precision_score,recall_score,f1_score

accuracy=accuracy_score(ytrain,y_pred_train)
accuracytest=accuracy_score(ytest,y_pred_test)
print("accuracy train score:",accuracy)
print("accuracy test score :",accuracytest)
print()

recall=recall_score(ytrain,y_pred_train)
recalltest=recall_score(ytest,y_pred_test)
print("recall  train score :",recall)
print("recall test score   :",recalltest)
print()

precision=precision_score(ytrain,y_pred_train)
precisiontest=precision_score(ytest,y_pred_test)
print("precision  train score:",precision)
print("precision test score  :",precisiontest)
print()

f1=f1_score(ytrain,y_pred_train)
f1_scoretest=f1_score(ytest,y_pred_test)
print("f1_score train score  :",f1)
print("f1_score test score   :",f1_scoretest)
print()
```

## Task 5: Visualization of SVM Results
1.Visualize classification results on the testing data.

```python
#from the confusion matrics we can visualize the result

from sklearn.metrics import confusion_matrix

sns.heatmap(confusion_matrix(ytest,y_pred_test),annot=True,cmap="Blues",fmt='d')
plt.xlabel("predicted")
plt.ylabel("actual")
plt.title("Confusion matrix")
plt.show()
```

```python
#apply pca(principle components analysis) to reduce the dimensions
from sklearn.decomposition import PCA
from sklearn.svm import SVC

pca=PCA(n_components=2)                  #creating pca instance with n_components 2
xtrain_pca=pca.fit_transform(xtrain)         #fitting pca on xtrain data
xtest_pca=pca.transform(xtest)               #fitting the pca on test data

modelsvm=SVC(kernel='rbf',gamma=5,probability=True)     #creating a model of svc
modelsvm.fit(xtrain_pca,ytrain)                      #fitting the model on xtrain pca and ytrain

y_pred=modelsvm.predict(xtest_pca)     
```

```python
plt.figure(figsize=(8,6))
plt.scatter(xtest_pca[:,0],xtest_pca[:,1],c=y_pred,cmap="coolwarm",edgecolors="k",alpha=0.6)
plt.title("SVM Classification Results (PCA-Reduced Mushroom dataset)")
plt.xlabel("Principle component 1")
plt.ylabel("principle component 2")
plt.colorbar(label='class')
plt.show()
```

## Task 6: Parameter Tuning and Optimization
1.Experiment with different SVM hyperparameters (e.g., kernel type, regularization parameter) to optimize performance.

```python
#experiment with different regularization parameters



svm=SVC(kernel='linear',C=3)

svm.fit(xtrain,ytrain)

y_pred_train=svm.predict(xtrain)
y_pred_test=svm.predict(xtest)

accuracy=accuracy_score(ytrain,y_pred_train)
accuracytest=accuracy_score(ytest,y_pred_test)

print("accuracy score train",accuracy)
print("accuracy score test",accuracytest)
```

## Apply grid search cv for best hyperparameter

```python
#experiment with different svm hyperparemeters kernal types

from sklearn.model_selection import GridSearchCV

param_grids={"kernel":['linear'],
                    "C":[1,2,3]}

svm=SVC()

gridsearch=GridSearchCV(svm,param_grids,n_jobs=-1)

gridsearch.fit(xtrain,ytrain)

y_pred_train=gridsearch.predict(xtrain)
y_pred_test=gridsearch.predict(xtest)

accuracy=accuracy_score(ytrain,y_pred_train)
accuracytest=accuracy_score(ytest,y_pred_test)

print("accuracy score train",accuracy)
print("accuracy score test",accuracytest)

gridsearch.best_params_,gridsearch.best_score_
```

```python
#applying the kernal type as polynomial

param_grid={"kernel":['poly'],
           'degree':[1,2,3]}


svm=SVC()

grid=GridSearchCV(svm,param_grids,n_jobs=-1)

grid.fit(xtrain,ytrain)

y_pred_train=grid.predict(xtrain)
y_pred_test=grid.predict(xtest)

accuracy=accuracy_score(ytrain,y_pred_train)
accuracytest=accuracy_score(ytest,y_pred_test)

print("accuracy score train",accuracy)
print("accuracy score test",accuracytest)

grid.best_params_,grid.best_score_
```

```python
#applying the kernal type as polynomial

param_grid={"kernel":['rbf'],
           'degree':[1,2,3]}


svm=SVC()

grid=GridSearchCV(svm,param_grids,n_jobs=-1)

grid.fit(xtrain,ytrain)

y_pred_train=grid.predict(xtrain)
y_pred_test=grid.predict(xtest)

accuracy=accuracy_score(ytrain,y_pred_train)
accuracytest=accuracy_score(ytest,y_pred_test)

print("accuracy score train",accuracy)
print("accuracy score test",accuracytest)

grid.best_params_,grid.best_score_
```
